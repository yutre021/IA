```python
 import cv2
import matplotlib.pyplot as plt
import numpy as np

#carrega a imagem
imagem = cv2.imread('tomografia_concreto.jpg', cv2.IMREAD_GRAYSCALE)
lg,cp=imagem.shape

#definir a faixa de pixels desejada
limite_inferior = 130
limite_superior = 255

#gerear uma mascara binaria para os pixels
mascara = np.logical_and(imagem >= limite_inferior, imagem <= limite_superior )
print(mascara)
plt.imshow(mascara, cmap="gray")

#contar os pixels brancos na mascara binaria do valor TRUE
quantidade_pixels_faixa = np.sum(mascara)
print(f"concentraçao: {100*quantidade_pixels_faixa/(lg*cp):,.2f}%")

```

```python

import cv2
import matplotlib.pyplot as plt
import numpy as np

#carrega a imagem
imagem = cv2.imread('tomografia_concreto.jpg')

#definir a faixa de pixels desejada
limite_inferior = 130
limite_superior = 200

#converte a imagem colorida para tons de cinza
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

#gerar uma mascara binaria para os pixels dentro da faixa
mascara = np.logical_and(imagem_gray >= limite_inferior, imagem_gray <= limite_superior )
print(mascara)

#criar uma cópia da imagem original para desenhar sobre ela
imagem_colorida = imagem.copy()

#Definir a cor amarela em (RGB)
cor_amarela = (255,255,0)

#aplicar a cor amarela apenas nos pixels dentro da faixa especificada
imagem_colorida[mascara] = cor_amarela
```
