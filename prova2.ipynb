```python
#Yuri Dlouhy - 767447
import cv2
import matplotlib.pyplot as plt
import numpy as np
# Carregar a imagem
imagem = cv2.imread('C:/Users/767447/Desktop/Processamento_Im/folha_zinco.jpg')
plt.imshow(imagem)
plt.axis('off')  # Desligar os eixos
plt.show()

# Extrair os canais de cores (BGR)
canal_azul, canal_verde, canal_vermelho = cv2.split(imagem)

# Definir as porcentagens desejadas para cada canal
porcentagem_azul = 0.875
porcentagem_verde = 0.845
porcentagem_vermelho = 0

# Ajustar a intensidade de cada canal de acordo com as porcentagens
canal_azul = np.clip(canal_azul * porcentagem_azul, 0, 255).astype(np.uint8)
canal_verde = np.clip(canal_verde * porcentagem_verde, 0, 255).astype(np.uint8)
canal_vermelho = np.clip(canal_vermelho * porcentagem_vermelho, 0, 255).astype(np.uint8)

# Mesclar os canais para formar a nova imagem
imagem_colorida = cv2.merge([canal_azul, canal_verde, canal_vermelho])

# Exibir a imagem com as novas porcentagens de cores
plt.imshow(cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Desligar os eixos
plt.show()


cv2.imwrite('C:/Users/767447/Desktop/Processamento_Im/zinco.png', imagem_colorida)
lower_yellow = np.array([255, 255, 0], dtype = "uint8")
upper_yellow= np.array([255, 255, 0], dtype = "uint8")

mask = cv2.inRange(imagem, lower_yellow, upper_yellow)

detected_output = cv2.bitwise_and(imagem, imagem, mask = mask)
plt.imshow(detected_output)
```

```python
#Define as cordenadas do ROI (Retengulo)
x1, y1 = 200, 200 #canto superior esquerdo
x2, y2 = 400, 350 # canto inferior direito

#Recorte do ROI
roi = imagem_colorida[x1:x2, y1:y2]

#print(roi.shape)
plt.imshow(roi, cmap= 'gray')
```
